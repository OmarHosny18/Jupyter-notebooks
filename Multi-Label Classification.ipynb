```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.multioutput import MultiOutputClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Algorithms
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit

# Metrics
from sklearn.metrics import (
    classification_report, 
    confusion_matrix, 
    roc_auc_score, 
    roc_curve, 
    hamming_loss
)

# Cross-validation
from sklearn.model_selection import StratifiedKFold, cross_validate

# Load Selected Dataset
X = pd.read_csv('X_selected_no_crix.csv')
y = pd.read_csv('Y_selected_no_crix.csv')

# Smell columns for multi-label classification
smell_columns = [
    'Brain Class', 'Data Class', 
    'God Class', 'Schizofrenic Class', 
]

# Use MultilabelStratifiedShuffleSplit instead of train_test_split
msss = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_idx, test_idx in msss.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
# Define classifiers with improved configurations for reducing overfitting
classifiers = {
    'Random Forest': MultiOutputClassifier(RandomForestClassifier(
        n_estimators=100,
        max_depth=10,  # Limit tree depth
        min_samples_split=5,  # Minimum samples required to split
        min_samples_leaf=2,  # Minimum samples required at leaf node
        max_features='sqrt',  # Use sqrt of features for each split
        class_weight="balanced",
        random_state=42
    )),
    'Gradient Boosting': MultiOutputClassifier(GradientBoostingClassifier(
        n_estimators=100,
        max_depth=6,  # Shallow trees to prevent overfitting
        learning_rate=0.1,  # Lower learning rate for better generalization
        min_samples_split=5,
        subsample=0.8,  # Use 80% of samples for each tree
        random_state=42
    )),
    'XGBoost': MultiOutputClassifier(XGBClassifier(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        min_child_weight=3,  # Minimum sum of instance weight in a child
        subsample=0.8,
        colsample_bytree=0.8,  # Use 80% of features for each tree
        reg_alpha=0.1,  # L1 regularization
        reg_lambda=1.0,  # L2 regularization
        use_label_encoder=False,
        random_state=42
    )),
    'Logistic Regression': MultiOutputClassifier(LogisticRegression(
        multi_class='ovr',
        class_weight='balanced',
        max_iter=1000,
        C=1.0,  # Inverse of regularization strength
        penalty='l2',  # L2 regularization
        solver='saga',  # Handles L1 and L2 regularization
        random_state=42
    )),
    'K-Nearest Neighbors': MultiOutputClassifier(KNeighborsClassifier(
        n_neighbors=5,
        weights='distance',  # Weight points by inverse of their distance
        metric='minkowski',
        p=2,  # Euclidean distance
        leaf_size=30
    ))
}

# Enhanced evaluation function with k-fold cross-validation
def enhanced_model_evaluation(model, X_train, X_test, y_train, y_test, model_name):
    """
    Perform comprehensive model evaluation with stratified k-fold cross-validation
    by using a combined label approach
    """
    print(f"\n{'='*50}")
    print(f"Evaluating {model_name}")
    print(f"{'='*50}")
    
    # Create a combined label for stratification
    # Convert multi-label to single label by treating each combination as a unique class
    y_combined = y_train.apply(lambda x: ''.join(x.astype(str)), axis=1)
    
    # # Initialize StratifiedKFold
    n_splits = 5
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    # # Arrays to store cross-validation scores
    cv_scores = {label: [] for label in smell_columns}
    cv_hamming = []
    
    print("\nPerforming Stratified Cross-validation...")
    
    # # Perform stratified cross-validation
    for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, y_combined), 1):
        print(f"\nFold {fold}/{n_splits}")
        
    #     # Split data
        X_cv_train = X_train.iloc[train_idx]
        X_cv_val = X_train.iloc[val_idx]
        y_cv_train = y_train.iloc[train_idx]
        y_cv_val = y_train.iloc[val_idx]
        
    #     # Fit and predict
        model.fit(X_cv_train, y_cv_train)
        y_cv_pred = model.predict(X_cv_val)
        
    #     # Calculate metrics for each label
        for idx, label in enumerate(smell_columns):
            fold_report = classification_report(
                y_cv_val.iloc[:, idx],
                y_cv_pred[:, idx],
                output_dict=True
            )
            cv_scores[label].append(fold_report['weighted avg']['f1-score'])
        
    #     # Calculate hamming loss for this fold
        fold_hamming = hamming_loss(y_cv_val, y_cv_pred)
        cv_hamming.append(fold_hamming)
    
    # Print cross-validation results
    print("\nCross-validation Results:")
    for label in smell_columns:
        scores = cv_scores[label]
        print(f"\n{label}:")
        print(f"Mean F1-Score: {np.mean(scores):.4f} (+/- {np.std(scores) * 2:.4f})")
    
    # print(f"\nMean Hamming Loss: {np.mean(cv_hamming):.4f} (+/- {np.std(cv_hamming) * 2:.4f})")
    
    # Fit the model on the full training set
    model.fit(X_train, y_train)
    
    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Calculate and store metrics
    results = {
        'model_name': model_name,
        'label_performance': {},
        'hamming_loss': {
            'train': hamming_loss(y_train, y_train_pred),
            'test': hamming_loss(y_test, y_test_pred)
        },
         'cv_scores': cv_scores,
         'cv_hamming': cv_hamming
    }
    
    # Evaluate each label
    for idx, label_name in enumerate(smell_columns):
        print(f"\nDetailed Performance for {label_name}")
        
        # Training metrics
        train_report = classification_report(
            y_train.iloc[:, idx],
            y_train_pred[:, idx],
            output_dict=True
        )
        
        # Testing metrics
        test_report = classification_report(
            y_test.iloc[:, idx],
            y_test_pred[:, idx],
            output_dict=True
        )
        
        # Store results
        results['label_performance'][label_name] = {
            'train': train_report,
            'test': test_report
        }
        
        # Print reports
        print("\nTraining Performance:")
        print(pd.DataFrame(train_report).T)
        print("\nTesting Performance:")
        print(pd.DataFrame(test_report).T)
      
        # Check for overfitting
        train_f1 = train_report['weighted avg']['f1-score']
        test_f1 = test_report['weighted avg']['f1-score']
        if train_f1 - test_f1 > 0.1:
            print(f"\nWarning: Possible overfitting detected for {label_name}")
            print(f"Training F1: {train_f1:.4f}, Testing F1: {test_f1:.4f}")

    
      # Overall classification reports
    print("\n--- Overall Classification Report (Training Data) ---")
    print(classification_report(y_train, y_train_pred, target_names=smell_columns))
    print("\n--- Overall Classification Report (Testing Data) ---")
    print(classification_report(y_test, y_test_pred, target_names=smell_columns))
         
    return results

# Store results for comparison
all_results = {}

# Evaluate each classifier
for name, clf in classifiers.items():
   
    pipeline = clf

    
    try:
        # Comprehensive evaluation
        results = enhanced_model_evaluation(
            pipeline,
            X_train, X_test,
            y_train, y_test,
            name
        )
        
        all_results[name] = results
        
    except Exception as e:
        print(f"\nError evaluating {name}:")
        print(str(e))
        continue

# Visualize results for successful models
if all_results:
    plt.figure(figsize=(12, 6))
    models = list(all_results.keys())
    train_losses = [results['hamming_loss']['train'] for results in all_results.values()]
    test_losses = [results['hamming_loss']['test'] for results in all_results.values()]

    x = np.arange(len(models))
    width = 0.35

    plt.bar(x - width/2, train_losses, width, label='Training Hamming Loss')
    plt.bar(x + width/2, test_losses, width, label='Testing Hamming Loss')

    plt.xlabel('Models')
    plt.ylabel('Hamming Loss')
    plt.title('Hamming Loss Comparison (Training vs Testing)')
    plt.xticks(x, models, rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()
    plt.figure(figsize=(15, 10))

# Metrics to plot


# Print summary of cross-validation results
print("\n--- Cross-Validation Summary ---")
for model, results in all_results.items():
    print(f"\n{model}:")
    for metric, value in results['cv_scores'].items():
        print(f"  {metric}: Mean={np.mean(value):.4f}, Std={np.std(value):.4f}")
```

```python
import pandas as pd
import pickle
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report, hamming_loss
from sklearn.model_selection import StratifiedKFold
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit

# Load data
X = pd.read_csv('X_selected_no_crix.csv')
y = pd.read_csv('Y_selected_no_crix.csv')

# Use MultilabelStratifiedShuffleSplit instead of train_test_split
msss = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_idx, test_idx in msss.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# Define KNN Model
knn = MultiOutputClassifier(KNeighborsClassifier(
    n_neighbors=5,
    weights='distance',
    metric='minkowski',
    p=2,
    leaf_size=30
))

# Define Pipeline
pipeline = Pipeline([
    ('knn', knn) 
])

# Fit the model
pipeline.fit(X_train, y_train)

# Save the pipeline
with open('MyKNNModel.pkl', 'wb') as f:
    pickle.dump(pipeline, f)

# Load pipeline
with open('MyKNNModel.pkl', 'rb') as f:
    loaded_pipeline = pickle.load(f)

print("Pipeline loaded successfully")

# Check class distribution
print("\nTraining set label distribution:")
print(y_train.mean())
print("\nTest set label distribution:")
print(y_test.mean())

# Enhanced Model Evaluation
def enhanced_model_evaluation(model, X_train, X_test, y_train, y_test, model_name):
    """
    Perform comprehensive model evaluation using Stratified K-Fold
    and calculate classification reports.
    """
    print(f"\n{'='*50}")
    print(f"Evaluating {model_name}")
    print(f"{'='*50}")
    
    


    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    print("\n--- Overall Classification Report (Training Data) ---")
    print(classification_report(y_train, y_train_pred))

    print("\n--- Overall Classification Report (Testing Data) ---")
    print(classification_report(y_test, y_test_pred))

    # Return evaluation results
    return {
        'model_name': model_name,
        'hamming_loss': {
            'train': hamming_loss(y_train, y_train_pred),
            'test': hamming_loss(y_test, y_test_pred)
        },
      
    }

# Run evaluation
evaluation_results = enhanced_model_evaluation(loaded_pipeline, X_train, X_test, y_train, y_test, "KNN Model")
print("Features expected by model:", pipeline.named_steps['classifier'].feature_names_in_)

```
